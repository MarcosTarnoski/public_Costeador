```python
xl = pd.ExcelFile("compras.xlsx")
df_main = pd.DataFrame()
for sheet in xl.sheet_names:
    print(sheet)
    if "aux" not in sheet:
        df = pd.read_excel(xl, sheet)
        df_main = pd.concat([df_main, df])
        
df_main.to_excel("compras_consolidado.xlsx")
```
